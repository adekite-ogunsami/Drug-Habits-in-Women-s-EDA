HANDLING MISSING VALUES AND EXPLANATIONS

```python
missing_counts = df.isnull().sum().sort_values(ascending=False)

missing_percent = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percent Missing': missing_percent
})

missing_summary.head(33)
```

THERE ARE 7 COLUMNS IN THE DATASET WITH MISSING VAUES AND FOR DIFFERENT REASONS

```python
df_cl = df.copy()
```

DROPPING ROWS WITH VALUES MISSING IN THESE COLUMNS

```python

# Drop only rows missing any of the *three* essential variables
required_columns = ['frdmjmon', 'talkprob', 'irmaritstat']
df_cleaned = df_cl.dropna(subset=required_columns)
```

IMPUTE MISSING VALUES WITH MEDIAN METHOD

```python
poverty_mean = df_cleaned['poverty3'].median()

df_cleaned['poverty3'] = df_cleaned['poverty3'].fillna(poverty_mean)
```
